```python
# Vehicles reached 20000 total mileage
max_mileage_per_vehicle = df_sorted.groupby('vehicle_id')['total_mileage'].max()

bikes_above_20000_per_vehicle = max_mileage_per_vehicle[max_mileage_per_vehicle >= 20000]

num_bikes_above_20000_per_vehicle = len(bikes_above_20000_per_vehicle)

print(f"20,000 or more: {num_bikes_above_20000_per_vehicle}")
print(f"less than 20,000: {df_sorted['vehicle_id'].nunique()-num_bikes_above_20000_per_vehicle}")
```

```python
# for bikes above 20,000 how many drops
filtered_rows = df_sorted.copy()
filtered_rows = filtered_rows[filtered_rows['vehicle_id'].isin(bikes_above_20000_per_vehicle.index)]

filtered_rows['mileage_diff'] = filtered_rows.groupby('vehicle_id')['total_mileage'].diff()

count_drops_below_20000 = filtered_rows[filtered_rows['mileage_diff'] < 0]['vehicle_id'].nunique()
count_no_drops_below_20000 = len(bikes_above_20000_per_vehicle) - count_drops_below_20000

print(f"dropping below 20,000: {count_drops_below_20000}")
print(f"not dropping below 20,000: {count_no_drops_below_20000}")
```

# Removing rows of vehicles with noise in the total mileage, battery usage and less than 30 records

### handling battery data *issues*

```python

def clean_battery_serial(serial):
    if str(serial)[0].isdigit() and len(str(serial)) == 40:
        return serial
    elif str(serial)[0].isalpha() and len(str(serial)) >= 17:
        return str(serial)[:17]
    else:
        return serial

df_sorted_2=df_sorted.copy()
df_sorted_2['battery_serial_number'] = df_sorted_2['battery_serial_number'].apply(clean_battery_serial)

df_sorted_2[['vehicle_id','battery_serial_number']]
```

### removing rows where 'total_mileage' is less than the previous one for a unique 'vehicle_id'


```python
# removing rows where 'total_mileage' is less than the previous one for a unique 'vehicle_id'
df_filtered = df_sorted_2.copy()

df_filtered['cumulative_max'] = df_filtered.groupby('vehicle_id')['total_mileage'].cummax()
df_filtered = df_filtered[df_filtered['total_mileage'] >= df_filtered['cumulative_max']]

df_filtered.drop(columns=['cumulative_max'], inplace=True)

print("Filtered DataFrame:")
df_filtered
```

### Handleling vehicles with extreme battery usage

```python
# Handleling vehicles with extreme battery usage
from scipy.stats import zscore
battery_counts = df_filtered_2.groupby('vehicle_id')['battery_serial_number'].nunique().reset_index()
battery_counts['z_score'] = zscore(battery_counts['battery_serial_number'])

z_score_threshold = 3

outliers = battery_counts[abs(battery_counts['z_score']) > z_score_threshold]['vehicle_id']

df_filtered_3 = df_filtered_2[~df_filtered_2['vehicle_id'].isin(outliers)]

df_filtered_3
```
